```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold
```

```python
santandar_data = pd.read_csv("train.csv", nrows=15000)
```

```python
santandar_data.head()
```

```python
s_data = santandar_data.drop("TARGET", axis =1)
t_data = santandar_data["TARGET"]
```

```python
s_data.shape
```

# You have to further split data into train and test 

```python
x_train,x_test,y_train,y_test = train_test_split(s_data,t_data,test_size = 0.2,random_state = 42)
```

# Use train data to identify the constant

```python
constant_filter = VarianceThreshold(threshold =0)
```

```python
constant_filter.fit(x_train)
```

```python
constant_filter.get_support().sum()
```

```python
con_list=[not temp for temp in constant_filter.get_support()]
con_list
```

```python
cons_column = s_data.columns[con_list]
```

# List down all identified constant features  

```python
cons_column
```

```python
len(cons_column)
```

# Drop all constant features from training set

```python
x_train.drop(labels=cons_column, axis=1, inplace=True)
```

```python
x_train.shape
```

# Use train data to identify the Quasi  constant features

```python
quasi_constant_filter = VarianceThreshold(threshold= 0.01)
```

```python
quasi_constant_filter.fit(x_train)
```

```python
quasi_constant_filter.get_support().sum()
```

```python
quasi_con_list=[not temp for temp in quasi_constant_filter.get_support()]
quasi_con_list

```

```python
quasi_cons_column = x_train.columns[quasi_con_list]
```

```python
quasi_cons_column
```

```python
len(quasi_cons_column)
```

```python
x_train_quasi = quasi_constant_filter.transform(x_train)
```

```python
x_train_quasi.shape,x_train.shape
```

```python
x_train_output = pd.DataFrame(x_train_quasi)
```
